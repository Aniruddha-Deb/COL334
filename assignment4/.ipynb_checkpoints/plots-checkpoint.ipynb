```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data_files = [f for f in os.listdir('Q1') if f.endswith(".cwnd")]

dfs = {f: pd.read_csv(f"Q1/{f}", delim_whitespace=True, header=None, names=['Time', 'StartWnd', 'EndWnd']) for f in data_files}
```

```python
for fname,df in dfs.items():
    plt.figure(figsize=(20,5), dpi=100)
    plt.cla()
    plt.plot(df['Time'], df['EndWnd'])
    title = fname.split('.')[0].split('_')
    title = f"{title[1]} ({title[2]})"    
    plt.title(f"Congestion window size for TCP {title}")
    plt.xlabel("Time (s)")
    plt.ylabel("Cwnd")
    plt.xlim(0,30)
    plt.savefig(f"Q1/cwnd_plot_{title}.pdf", bbox_inches="tight")
```

```python
conv = pd.read_csv('Q2/convergence.csv')

plt.figure(figsize=(10,5), dpi=100)
plt.xscale('symlog')
plt.plot(conv['Time'], conv['100']-0.075, label="100 ms", marker='o')
plt.plot(conv['Time'], conv['1000']+0.05-0.075, label="1000 ms", marker='o')
plt.plot(conv['Time'], conv['20000']+0.1-0.075, label="20000 ms", marker='o')
plt.plot(conv['Time'], conv['80000']+0.15-0.075, label="80000 ms", marker='o')
plt.legend()
plt.title('Convergence v/s Time')
plt.xlabel('Time (s)')
plt.ylabel('Converged')
plt.ylim(-0.2,1.8)
plt.savefig('Q2/q2a_convergence.pdf', bbox_inches='tight')
```
